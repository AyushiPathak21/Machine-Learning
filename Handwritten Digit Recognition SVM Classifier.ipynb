```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_digits
```

```python
data=load_digits()
```

```python
data
```

```python
print(data.data)
print(data.target)
print(data.data.shape)
print(data.images.shape)
DataImageLength=len(data.images)
print(DataImageLength)

```

```python
n=1
import matplotlib.pyplot as plt
plt.gray()
plt.matshow(data.images[n])
plt.show()

data.images[n]
```

```python
x=data.images.reshape((DataImageLength,-1))
x
```

```python
y=data.target
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn import svm
model=svm.SVC(kernel='linear')
model.fit(x_train,y_train)
```

```python
n=5
result=model.predict(data.images[n].reshape((1,-1)))
plt.imshow(data.images[n],cmap=plt.cm.gray_r,interpolation='nearest')
print(result)
print("\n")
plt.axis('off')
plt.title('%i' %result)
plt.show()
```

```python
y_pred=model.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy of the Model: {0}%".format(accuracy_score(y_test,y_pred)*100))
```

```python
from sklearn import svm

model1=svm.SVC(kernel='linear')
model2=svm.SVC(kernel='rbf')
model3=svm.SVC(gamma=0.001)

model1.fit(x_train,y_train)
model2.fit(x_train,y_train)
model3.fit(x_train,y_train)

y_predModel1=model1.predict(x_test)
y_predModel2=model1.predict(x_test)
y_predModel3=model1.predict(x_test)


print("Accuracy of the Model: {0}%".format(accuracy_score(y_test,y_predModel1)*100))
print("Accuracy of the Model: {0}%".format(accuracy_score(y_test,y_predModel2)*100))
print("Accuracy of the Model: {0}%".format(accuracy_score(y_test,y_predModel3)*100))
```
